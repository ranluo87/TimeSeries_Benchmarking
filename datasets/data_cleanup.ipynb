```python
import numpy as np
import pandas as pd
import os
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('TkAgg')

dataset_dir = "C:\Python Projects\TimeSeries_Benchmarking\datasets\Select"

for file in os.listdir(dataset_dir):
    if not file.endswith(".csv"):
        continue

    data_path = os.path.join(dataset_dir, file)
    methane_df = pd.read_csv(data_path, parse_dates=True, index_col=0)

    largest_gap = methane_df.index.diff().max()
    print("Largest Gap for {} is {}".format(file.split(".")[0], largest_gap))
    methane_df = methane_df.asfreq('h').reset_index()
    methane_df = methane_df.ffill()

    print('test')
```
