```python
import pandas as pd
import matplotlib.pyplot as plt

import os
```

```python
result_dir = './GRU'

fig, axes = plt.subplots(2, 2, figsize=(20, 12))

axes = axes.ravel()
idx = 0

for file in os.listdir(result_dir):
    if 'Test Fitting' not in file:
        continue
    
    test_fitting_path = os.path.join(result_dir, file)
    test_fitting_df = pd.read_csv(test_fitting_path, parse_dates=True, index_col=0)
    
    axes[idx].plot(test_fitting_df.index, test_fitting_df['pred'])
    axes[idx].plot(test_fitting_df.index, test_fitting_df['true'])  
    
    idx += 1

plt.show()
    
```
